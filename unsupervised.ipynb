```python
# Install necessary libraries (if not already installed)
!pip install -q scikit-learn pandas matplotlib

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Create a simple dataset (for example: study hours vs marks)
data = {
    'hours_studied': [1, 2, 3, 8, 9, 10, 2, 3, 9, 10],
    'marks_obtained': [20, 25, 30, 85, 88, 95, 28, 35, 80, 92]
}

df = pd.DataFrame(data)

# View data
print(df)

# Convert to numpy array
X = df.values

# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)  # We assume we want 2 groups
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Add labels to the DataFrame
df['Cluster'] = labels

# Print the new DataFrame
print("\nClustered Data:")
print(df)

# Plot the clusters
plt.scatter(df['hours_studied'], df['marks_obtained'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('KMeans Clustering (Unsupervised Learning)')
plt.show()
```
