```python
# Install required libraries
!pip install -q langchain-google-genai langchain

# Import libraries
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import os

# Set Gemini API Key
os.environ["GOOGLE_API_KEY"] = "AIzaSyDrlaYZ07flQD26CKcq_a2kA5VYY2dv4Lc"

# Create a Gemini LLM object
llm = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0.7)

# Create a prompt template
prompt = PromptTemplate(
    input_variables=["topic"],
    template="Write a short inspiring story about {topic}."
)

# Create an LLMChain
chain = LLMChain(llm=llm, prompt=prompt)

# Run the chain with a topic
response = chain.run("never giving up")

print("Response:\n", response)
```
